```python
def m_cd(data):
    m_cdW = []
    m_cdR = []

    i= 0
    while i < len(data["CrW"]):
        val = cf_CrW[i] + cf_NiW[i] + cf_CuW[i] + cf_AsW[i] + cf_CdW[i] + cf_PbW[i]  
        val = val/6
        m_cdW.append(val)

        val = cf_CrR[i] + cf_NiR[i] + cf_CuR[i] + cf_AsR[i] + cf_CdR[i] + cf_PbR[i]  
        val = val/6
        m_cdR.append(val)

        i+=1
    cf["m_cdW"], cf["m_cdR"] = m_cdW, m_cdR
    return cf
```

```python
cf = m_cd(cf)
cf
```

```python
mhq_Cr = []
mhq_Ni = []
mhq_Cu = []
mhq_As = []
mhq_Cd = []
mhq_Pb = []

i = 0
while i < len(CW.Cr):
    val = math.sqrt(CW.Cr[i] * ((1/TEL["Cr"]) + (1/PEL["Cr"]) + (1/SEL["Cr"])))
    mhq_Cr.append(val)
    
    val = math.sqrt(CW.Ni[i] * ((1/TEL["Ni"]) + (1/PEL["Ni"]) + (1/SEL["Ni"])))
    mhq_Ni.append(val)
    
    val = math.sqrt(CW.Cu[i] * ((1/TEL["Cu"]) + (1/PEL["Cu"]) + (1/SEL["Cu"])))
    mhq_Cu.append(val)
    
    val = math.sqrt(CW.As[i] * ((1/TEL["As"]) + (1/PEL["As"]) + (1/SEL["As"])))
    mhq_As.append(val)
    
    val = math.sqrt(CW.Cd[i] * ((1/TEL["Cd"]) + (1/PEL["Cd"]) + (1/SEL["Cd"])))
    mhq_Cd.append(val)

    val = math.sqrt(CW.Pb[i] * ((1/TEL["Pb"]) + (1/PEL["Pb"]) + (1/SEL["Pb"])))
    mhq_Pb.append(val)

    i +=1
    
mhq_Cd, mhq_Ni
```
